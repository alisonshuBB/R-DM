```r
#explore dataset structure
data('iris')
class(iris)
dim(iris)
str(iris)
head(iris,n=5) #tail(iris,n=5)
summary(iris)
```

```r
#explore specific numeric variance
iris$Sepal.Length[1:10] #iris[1:10,"Sepal.Length"]
quantile(iris$Sepal.Length) #var(iris$Sepal.Length)
hist(iris$Sepal.Length)
plot(density(iris$Sepal.Length))
```

```r
#explore nominal variance
table(iris$Species)
pie(table(iris$Species))
barplot(table(iris$Species))
```

```r
# explore the relation between multiple variances
cov(iris$Sepal.Length,iris$Petal.Length)
cor(iris$Sepal.Length,iris$Petal.Length)

aggregate(Sepal.Length~Species,FUN=mean,data=iris) #FUN=var, percentile
boxplot(Sepal.Length~Species,data=iris)
with(iris,plot(Sepal.Length,Petal.Length,col=Species,pch=as.numeric(Species))) #jitter() to add some noise
```
